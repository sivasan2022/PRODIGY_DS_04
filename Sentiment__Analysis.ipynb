SENTIMENT ANALYSIS

Importing libraries

```python
#importing necessary libraries
import pandas as pd
import numpy as np
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

Impoting the dataset

```python
#importing the dataset

df=pd.read_csv('Tweets.csv')
df
```

Column analysis

```python
#analysing the columns
print(df.columns)
```

```python
#printing the head
print(df.head())
```

PREPROCESSING THE DATA

```python
#preprocessing the data
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
data=df.apply(LabelEncoder().fit_transform)
print(data.head())
```

DOWNLOADING Vader_lexicon

```python
#downloading vader_lexicon and initializing SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
```

PERFORM SENTIMENT ANALYSIS

```python
# Perform sentiment analysis
for text in df:
    sentiment = sia.polarity_scores(text)
    compound_score = sentiment['compound']

    positive_count=0
    negative_count=0
    neutral_count=0

    if compound_score >= 0.05:
        result = "Positive"
        positive_count+=1
    elif compound_score <= -0.05:
        result = "Negative"
        negative_count+=1
    else:
        result = "Neutral"
        neutral_count+=1

    print(f"Text: {text}")
    print(f"Sentiment: {result} (Compound Score: {compound_score})")
    print()
```

INSIGHTS FROM SENTIMENT ANALYSIS

```python
#printing the insights obtained
print(f"Compound Score: {compound_score}")
print(f"Insights:")
print(f"Positive Tweets: {positive_count}")
print(f"Negative Tweets: {negative_count}")
print(f"Neutral Tweets: {neutral_count}")
```
